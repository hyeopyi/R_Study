```r
# 왼쪽값을 기초로한 계단 모양 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    xlim = c(min(y)-1, max(y)+1),  
    ylim = c(min(y)-1, max(y)+1),
    main = "Simple Time Series",
    type = "s"
    )
```

```r
# 오른쪽값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    xlim = c(min(y)-1, max(y)+1),  
    ylim = c(min(y)-1, max(y)+1),
    main = "Simple Time Series",
    type = "S"  # 대문자 S
    )
```

```r
# 오른쪽값을 기초로한 계단 모양 그래프 - 라인타입

y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    xlim = c(min(y)-1, max(y)+1),  
    ylim = c(min(y)-1, max(y)+1),
    main = "Simple Time Series",
    type = "S", 
    lty = 1
    )
```

```r
# 오른쪽값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    xlim = c(min(y)-1, max(y)+1),  
    ylim = c(min(y)-1, max(y)+1),
    main = "Simple Time Series",
    type = "p", 
    pch = c(18, 24, 9),
    col = c("red", "blue")
    )
```

```r
x <- runif(100)
y <- runif(100)

# pch : y > 0.5 이면 1, 아니면 18
# col : y > 0.5 이면 "red", 아니면 "blue"


plot(x, y,
     xlim = c(min(x)-0.1, max(x)+0.1),
     ylim = c(min(x)-0.1, max(x)+0.1),
     pch = ifelse(y>0.5, 1, 18),
     col = ifelse(y>0.5, "red", "blue")
    )
```

# 선그래프 

```r
# 3종류의 dataset의 선 그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = "AppleGothic")
options(warn =  -1)
plot(abc, 
     type ="o",
     col = "red",
     ylim = c(0, 400),
     axes = F, # x,y 좌표 Frame 제거,
     ann = F # x,y의 label 제거
    )

# x축과 x축의 라벨을 설정
axis(1, at=1:5, lab=c('서울', '대전', '대구', '광주', '원주'))

# y축과 y축의 라벨을 설정
axis(2, ylim=c(0,400))

# main title 지정
title(main='과일판매량', col.main="red") # plot 안에서 설정할경우 색은 검정으로 고정

# x축 title(label) 지정
title(xlab = "지역", col.lab = "blue")

# y축 title(label) 지정
title(ylab = "판매량", col.lab = "blue")

# 선그리기 함수로 def 그리기
lines(def, 
     type ="o",
     pch = 22,
     col = "green",
     lty = 2
    )

# 선그리기 함수로 ghi 그리기
lines(ghi, 
     type ="o",
     pch = 22,
     col = "blue",
     lty = 2
    )

# 범례
legend(4, 400,   # x, y 좌료
      c("야구장", "축구장", "해변가"),
      col = c("red", "green", "blue"),
      pch = c(21,22,22),
      lty = c(1,2,2),
      cex = 0.8
      
      )
```

# 파이차트

```r
sales <- c(210, 110, 400, 550, 600, 130)
pie(sales)
```

```r
par(family = "AppleGothic")
week <- c("월","화","수","목","금","토")
ratio <- sales/sum(sales)*100
label <- paste(week, '\n', ratio, "%")
pie(sales,
    main = "영업실적 차트",
    labels = label,
    col = rainbow(length(sales))   
   )
```

# 3D Pie Chart

```r
library(plotrix)
```

```r
par(family = "AppleGothic")
week <- c("월","화","수","목","금","토")
ratio <- sales/sum(sales)*100
label <- paste(week, '\n', ratio, "%")

pie3D(sales,
    main = "영업실적 차트",
    labels = label,  # color는 기본값이 rainbow
    explode = 0.05,
    labelcex =0.8
   )
```
