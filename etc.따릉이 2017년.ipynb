```r
check <- function(data) {
    data$대여소번호 <- str_remove_all(data$대여소번호, "'")
   
    rmlist <- c()
    for(i in 1:nrow(data)) {
        if(data$대여소번호[i] == '중랑센터'| data$대여소번호[i] == '상암센터 정비실' | data$대여소번호[i] == '위트콤'
          | data$대여소번호[i] == '위트콤공장'| data$대여소번호[i] == '중랑정비팀test 1005') {
            rmlist <- append(rmlist, i)
        }
    }
    data <- data[-rmlist,]
    data$대여소번호 <- as.numeric(data$대여소번호)
    return(sum(ifelse(data$대여소번호 > 2300, TRUE, FALSE)))
  
}
```

```r
data_refinement <- function(data) {
    # 대여소번호 숫자화
    data$대여소번호 <- str_remove_all(data$대여소번호, "'")
   
    rmlist <- c()
    for(i in 1:nrow(data)) {
        if(data$대여소번호[i] == '중랑센터'| data$대여소번호[i] == '상암센터 정비실' | data$대여소번호[i] == '위트콤'
          | data$대여소번호[i] == '위트콤공장'| data$대여소번호[i] == '중랑정비팀test 1005') {
            rmlist <- append(rmlist, i)
        }
    }
    data <- data[-rmlist,]
    data$대여소번호 <- as.numeric(data$대여소번호)
    
    # 대여일짜 데이트화
    data$대여일자 <- str_remove_all(data$대여일자, "'")
    data$대여일자 <- as.Date(data$대여일자, format = "%Y-%m-%d")
    
    # 대여시간 숫자화
    data$대여시간 <- str_remove_all(data$대여시간, "'")
    data$대여시간 <- as.numeric(data$대여시간)
    
    # 누적횟수 카운팅
    u_data <- unique(data)
    u_data$누적사용횟수 <- rep(0, nrow(u_data))
    
   for(i in 1:nrow(u_data)) {
       u_data$누적사용횟수[i] = sum(ifelse(u_data$대여일자[i]==data$대여일자 & u_data$대여시간[i]==data$대여시간 &
                         u_data$대여소번호[i]==data$대여소번호, 1, 0))
   }
    
    
    return(u_data)
    
}
```
